```python
import numpy as np
import matplotlib.pyplot as plt

# Given parameters
f0 = 1000  # Signal frequency (Hz)
fs = 8000  # Sampling frequency (Hz)

# Part (a): Spectrum of original signal
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.stem([-f0, f0], [2.5, 2.5], basefmt=' ')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('(a) Spectrum of Original Signal')
plt.xlim([-5000, 5000])
plt.ylim([0, 3])
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)

# Part (b): Spectrum of sampled signal (0 to 20 kHz)
plt.subplot(1, 2, 2)
frequencies = [1000, 7000, 9000, 15000, 17000]
magnitudes = [2.5, 2.5, 2.5, 2.5, 2.5]
plt.stem(frequencies, magnitudes, basefmt=' ')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('(b) Spectrum of Sampled Signal (0-20 kHz)')
plt.xlim([0, 20000])
plt.ylim([0, 3])
plt.grid(True, alpha=0.3)

# Mark sampling frequency multiples
for k in [1, 2]:
    plt.axvline(x=k*fs, color='gray', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```
